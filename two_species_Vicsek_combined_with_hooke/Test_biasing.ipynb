```python
#This list contains the different amount of proportions
p_list=np.linspace(0,0.5,10)

pos_with_different_amount_of_blue=np.zeros([len(p_list),iterations,N,2 ])
theta_with_different_amount_of_blue=np.zeros([len(p_list),iterations,N ])
for n,p in enumerate(p_list):
    s,k=Create_groups([ [ [1,v_0/10 ], [-1,v_0/10 ]  ],[[-1,v_0/10 ],[1,v_0/10 ]]   ],N,[p,1-p])
    pos,theta=Vicsek_hooke(v_0,eta,r, N,L,iterations,s,k,l)
    pos_with_different_amount_of_blue[n]=pos
    theta_with_different_amount_of_blue[n]=theta
np.save("Data/pos_with_different_amount_of_blue" ,pos_with_different_amount_of_blue)
np.save("Data/theta_with_different_amount_of_blue",theta_with_different_amount_of_blue)
```

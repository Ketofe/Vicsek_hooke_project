```python
import File_for_accesing_src
from Standard_parameters import *
from Vicsek_hooke import *
import numpy as np
import matplotlib.pyplot as plt
from Quiver import *
from tqdm import tqdm
```

```python
#Generate data
k_list=np.linspace(0.01,0.3,10)
pos_for_different_k=np.zeros([len(k_list),iterations,N,2 ])
theta_for_different_k=np.zeros([len(k_list),iterations,N ])
for n, q in tqdm(enumerate(k_list)):
    s,k=Create_groups([ [ [1,q ], [-1,q ]  ],[[-1,q ],[1,q ]]   ],N)
    pos,theta=Vicsek_hooke(v_0,eta,r, N,L,iterations,s,k,l)
    pos_for_different_k[n]=pos
    theta_for_different_k[n]=theta
np.save("Data/pos_with_different_k" ,pos_for_different_k)
np.save("Data/theta_with_different_k" ,theta_for_different_k)
```

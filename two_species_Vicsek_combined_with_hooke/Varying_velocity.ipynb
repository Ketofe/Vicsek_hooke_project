```python
import File_for_accesing_src
from Standard_parameters import *
from Vicsek_hooke import *
import numpy as np
import matplotlib.pyplot as plt
from Quiver import *
from tqdm import tqdm
```

```python
#Varying the velocity
v_list=np.linspace(0.01,0.3,10)
pos_for_different_v=np.zeros([len(v_list),iterations,N,2 ])
theta_for_different_v=np.zeros([len(v_list),iterations,N ])
for n, v in tqdm(enumerate(v_list)):
    pos,theta=Vicsek_hooke(v,eta,r, N,L,iterations,s,k,l)
    pos_for_different_v[n]=pos
    theta_for_different_v[n]=theta
np.save("Data/pos_with_different_v_0_01_to_0_3" ,pos_for_different_v)
np.save("Data/theta_with_different_v_0_01_to_0_3" ,theta_for_different_v)
```
